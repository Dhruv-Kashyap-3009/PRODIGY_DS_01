```python
#Getting the names of all the unique countries
print(df['Country Name'].unique())
print("\nThe total number of unique country names are:",df['Country Name'].nunique())
```

```python
#Getting the codes of all the unique countries
print(df['Country Code'].unique())
print("\nThe total number of unique country codes are:",df['Country Code'].nunique())
```

```python
#Filter population for total data
total_population_data=df[df['Indicator Code']=='SP.POP.TOTL']

#Sort data based on the total population for 2023
total_population_sorted= total_population_data.sort_values(by="2023",ascending=False)

#Get the top ten countries with the highest total population for 2023
total_top_ten_countries=total_population_sorted.head(10)

print("Total Population of top ten countries are:")
print(total_top_ten_countries[['Country Code']])
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="2023",y="Country Code",data=total_top_ten_countries,palette="coolwarm",hue="Country Code")
plt.title("Top ten countries of total population (2023)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()
```

```python
#Sort data based on the total population for 2022
total_population_sorted1=total_population_data.sort_values(by="2023",ascending=True)

#Get the bottom ten countries with the highest total population for 2023
total_bottom_ten_countries=total_population_sorted1.head(10)

print("Total Population of top ten countries are:")
print(total_bottom_ten_countries[['Country Code']])
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="2023",y="Country Code",data=total_bottom_ten_countries,palette="coolwarm",hue="Country Code")
plt.title("Bottom ten countries of total population (2023)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()
```
